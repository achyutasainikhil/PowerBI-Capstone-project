```python
# Local path to the existing repository
repo_path = 'achyutasainikhil/AnnualData'

# Open the existing repository
repo = git.Repo(repo_path)

# Pull changes from the remote repository
origin = repo.remotes.origin
origin.pull()

# Path to the file you want to extract
cheese_path = 'achyutasainikhil/AnnualData/Annual supply and utilization - Amer cheese, annual.csv'
butter_path = 'achyutasainikhil/AnnualData/Annual supply and utilization - Butter, annual.csv'
evap_path = 'achyutasainikhil/AnnualData/Annual supply and utilization - Evap cond whole, annual.csv'
lactose_path = 'achyutasainikhil/AnnualData/Annual supply and utilization - Lactose, annual.csv'
others_path = 'achyutasainikhil/AnnualData/Annual supply and utilization - Other cheese, annual.csv'

# Read the file into a DataFrame
cheese_path_df = pd.read_csv(cheese_path)
butter_path_df = pd.read_csv(butter_path)
evap_path_df = pd.read_csv(evap_path)
lactose_path_df = pd.read_csv(lactose_path)
others_path_df = pd.read_csv(others_path)

cheese_path_df['incredient'] = 'cheese'
butter_path_df['incredient'] = 'butter'
evap_path_df['incredient'] = 'evap'
lactose_path_df['incredient'] = 'lactose'
others_path_df['incredient'] = 'others'

merged_df = pd.concat([cheese_path_df, butter_path_df, evap_path_df, lactose_path_df, others_path_df], ignore_index=True)

merged_df.to_excel('merged_data.xlsx', index=False)
```
